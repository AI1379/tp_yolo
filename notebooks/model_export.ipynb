# 模型导出

由于 PyTorch 和 TensorFlow 之间有些依赖冲突，因此这个 notebook 需要使用一个单独的干净的环境给 TensorFlow 用。

```python
import os
import tensorflow as tf
# import onnx2tf
```

```python
def convert_onnx_to_tf_saved_model(onnx_model_path, tf_model_path):
    """Convert an ONNX to a TF SavedModel format by calling onnx2tf command line tool."""

    os.system(f"onnx2tf -i {onnx_model_path} -o {tf_model_path}")

    print(f"Converted {onnx_model_path} to {tf_model_path}")
```

```python
onnx_model_path = "E:/Projects/tp_yolo/runs/segment/train6/weights/best.onnx"
tf_model_path = "E:/Projects/tp_yolo/runs/segment/train6/weights/best_tf"
```

```python

convert_onnx_to_tf_saved_model(onnx_model_path, tf_model_path)
```

```python
def convert_tf_saved_model_to_tflite(saved_model_dir, tflite_model_path):
    """Convert a TF SavedModel to TFLite format."""
    converter = tf.lite.TFLiteConverter.from_saved_model(saved_model_dir)
    tflite_model = converter.convert()

    with open(tflite_model_path, "wb") as f:
        f.write(tflite_model)

    print(f"Converted {saved_model_dir} to {tflite_model_path}")
```

```python
tflite_model_path = (
    "E:/Projects/tp_yolo/runs/segment/train6/weights/best_float32.tflite"
)

convert_tf_saved_model_to_tflite(tf_model_path, tflite_model_path)
```
