```python
from ultralytics.models import YOLO

tflite_model = YOLO("../references/yolov8n-seg_saved_model/yolov8n-seg_float32.tflite")

results = tflite_model("../data/Test-TP.png")

results
```

```python
import matplotlib.pyplot as plt

def convert_to_rgb(image):
    return image[:, :, ::-1]

rgb = convert_to_rgb(results[0].plot())

plt.imshow(rgb)
```
